```python
import pandas as pd
import json
```

```python
file_path = '/content/교과서 데이터 정리본 계층형.xlsx'  # 파일 경로 수정 가능
df = pd.read_excel(file_path)

# JSON 형식으로 변환
json_data = []

for idx, row in df.iterrows():
    entry = {
        "content": row["본문"],
        "metadata": {
            "title": row["소주제"],
            "contentstype": "본문",
            "id": str(idx + 1),
            "source": "교과서"
        }
    }
    json_data.append(entry)

# JSON 파일로 저장
with open("교과서 데이터 정리본 계층형.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=2)
```
