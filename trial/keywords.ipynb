```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
import re
from gliner import GLiNER
from collections import defaultdict
from kiwipiepy import Kiwi
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
```

```python
file_path = "/content/1-6단원 판서+본문.xlsx"
df = pd.read_excel(file_path)
df.columns = ['topic', 'body', 'content']
```

```python
okt = Okt()
def extract_nouns(text):
    preprocessed = preprocess_text(text)
    nouns = okt.nouns(preprocessed)
    return ' '.join([n for n in nouns if len(n) > 1])
df['nouns_only'] = df['content'].apply(extract_nouns)
```

# TF-IDF

```python
# 전처리
def preprocess_text(text):
    text = str(text).replace('\n', ' ').replace('\r', ' ')
    text = re.sub(r'[\(\)\[\]\{\}0-9\-–—]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

# TF-IDF 계산
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['nouns_only'])

# 단어 목록
feature_names = vectorizer.get_feature_names_out()

# TF-IDF 기준 상위 키워드 추출
def get_top_keywords(row_index, top_n=5):
    row_data = tfidf_matrix[row_index].toarray().flatten()
    top_indices = row_data.argsort()[-top_n:][::-1]
    return [feature_names[i] for i in top_indices]
```

```python
for i in range(min(10, len(df))):
    keywords = get_top_keywords(i, top_n=5)
    print(f"소제목: {df.loc[i, 'topic']}")
    print(f"중심 키워드: {', '.join(keywords)}\n")
```

# NER

```python
# GLiNER 모델 로드
model = GLiNER.from_pretrained("urchade/gliner_multi")

# NER 라벨 정의
labels = ["Person", "Date", "Place", "Organization", "Case"]

# 후처리 함수 정의
def clean_entity(label, text):
    if label == "Person":
        return text.replace("을", "").replace("를", "")  # 조사 제거
    if label == "Place":
        return text.replace("에서", "").replace("의", "")  # 조사 제거
    if label == "Case":
        return text.replace("은", "").replace("는", "")  # 조사 제거
    return text

target_column = "content"


ner_results = []

for text in df[target_column].astype(str):  # 문자열로 변환
    try:
        entities = model.predict_entities(text, labels)
        entities_ref = defaultdict(list)
        for entity in entities:
            label = entity["label"]
            cleaned = clean_entity(label, entity["text"])
            if cleaned not in entities_ref[label]:
                entities_ref[label].append(cleaned)
        ner_results.append(dict(entities_ref))
    except Exception as e:
        ner_results.append({"error": str(e)})
```

# Key BERT

```python
model = SentenceTransformer("klue/roberta-base")  # 또는 다른 모델로 교체 가능
kw_model = KeyBERT(model=model)

texts = df["content"].astype(str)

# ✅ 3. 불용어 리스트 정의 (직접 추가 가능)
stopwords = set([
    "이", "그", "저", "것", "수", "등", "및", "더", "있다", "하였다", "했다", "위해", "대한", "으로", "하고", "에서",
    "으로서", "한다", "했다", "한다는", "되어", "된다", "같은", "있는", "없는", "하는", "하여", "하는", "되는"
])

# ✅ 4. 키워드 추출 함수 + 불용어 필터링
def extract_keywords(text):
    keywords = kw_model.extract_keywords(
        text,
        keyphrase_ngram_range=(1, 2),
        stop_words=None,  # KeyBERT stop_words는 사용 안 함
        top_n=10
    )
    # 불용어 필터링
    filtered = []
    for kw, score in keywords:
        # 단어 또는 구의 모든 단어가 stopwords에 포함되지 않으면 추가
        if not any(sw in kw for sw in stopwords):
            filtered.append(kw)
        if len(filtered) >= 5:
            break
    return filtered

# ✅ 5. 실행
df["klue"] = texts.apply(extract_keywords)
```
